### Column Descriptions

Source: DR1TOT_J Dataset Documentation

**SEQN**  
Unique respondent identifier.

**WTDRD1**  
Sample weight for Day 1 dietary recall. It indicates how many people in the U.S. population are represented by a given respondent. A value of `5.397605346934028e-79` (1,063 records) indicates that the Day 1 dietary recall was not completed or was incomplete.

**WTDRD2**  
Sample weight for Day 2 dietary recall. It indicates how many people in the U.S. population are represented by a given respondent. A value of `5.397605346934028e-79` (1,002 records) indicates that the Day 2 dietary recall was not completed or was incomplete.

**DR1DRSTZ**  
Indicates the reliability/status of the Day 1 dietary recall. Values 1, 2, 4, and 5 are present in the dataset. Value 4 corresponds to respondents on a breast milk diet and may need to be excluded from certain analyses.

**DR1EXMER**  
Identifier of the dietary interviewer who conducted the interview.

**DRABF**  
Indicator identifying infants who are breastfed.

**DRDINT**  
Indicates the number of dietary recall days collected (1 or 2).

**DR1DBIH**  
Number of days between the dietary recall day and the interview date. This variable may be useful for assessing potential recall bias, as a longer delay could increase the risk of inaccurate reporting.

**DR1DAY**  
Day of the week corresponding to the dietary recall:
- 1 = Sunday
- 2 = Monday
- 3 = Tuesday
- 4 = Wednesday
- 5 = Thursday
- 6 = Friday
- 7 = Saturday

**DR1LANG**  
Language used during the interview. See the documentation table for language codes.

**DR1MRESP**  
Identifies the primary respondent who answered the dietary interview questions. Interviews were often conducted in the participant's home, and family members could assist or answer on behalf of the respondent (e.g., infants or individuals with language barriers). Values 1–3 generally represent the most reliable response scenarios. See the documentation table for details.

**DR1HELP**  
Identifies who assisted the respondent during the interview. The cleanest scenario is value 12 ("No one"), indicating that the respondent answered independently without assistance. See the documentation table for details.

### Notes

- The `DR1DBIH` variable can be used to assess potential recall bias. The longer the delay between food consumption and the interview, the higher the risk of inaccurate reporting.
- The extremely small value `5.397605346934028e-79` in `WTDRD1` and `WTDRD2` is consistent with the NHANES documentation and indicates an incomplete or missing dietary recall rather than a data quality issue.
- Null values are consistent with the documentation and are a result of the survey design rather than data errors.

### Building silver layer

Description additional columns:

**RIAGENDER** - sex respondent
**RIDAGEYR** - respondent age in year  
**DR1TKCAL** - daily energy income by day_1 [kcal]  
**DR1TFIBE** - daily fiber income [g]  
**DR1TSODI** - daily sodium income [mg]

```python
silver_demo = demo[[
    "SEQN",
    "RIAGENDR",
    "RIDAGEYR"

]]

silver_day_1 = day_1[[
    "SEQN",
    "DR1DRSTZ",
    "DR1MRESP",
    "DR1HELP",
    "DR1TKCAL",
    "DR1TFIBE",
    "DR1TSODI"
]]

silver_df = silver_day_1.merge(silver_demo, on="SEQN", how="left")

silver_df["age_group"] = pd.cut(
    silver_df["RIDAGEYR"],
    bins=[0, 2, 19, 39, 65, 120],
    labels=["0-2", "3-19", "20-39", "40-65", "65+"]
)

# I determine flag for too low kcal < 800, too high kcal > 5500 and normal between 800 - 5500
silver_df.loc[silver_df["DR1TKCAL"] < 800, "kcal_flag"] = "low_kcal"

silver_df.loc[
    (silver_df["DR1TKCAL"] >= 800) & (silver_df["DR1TKCAL"] <= 5500),
    "kcal_flag"
] = "normal_kcal"

silver_df.loc[silver_df["DR1TKCAL"] > 5500, "kcal_flag"] = "high_kcal"
```

### Saving silver data frame to postgresql

```python
engine = create_engine(
    "postgresql://postgres:postgres@localhost:5432/diploma_db"
)

silver_df.to_sql(
    "nhanes_silver",
    engine,
    index=False,
    if_exists="replace"
)
```

### I define the parameters for the most reliable data

```python
quality_silver_data = silver_df[
    (silver_df["DR1DRSTZ"] == 1) &
    (silver_df["DR1MRESP"].isin([1, 2, 3])) &
    (silver_df["DR1HELP"] == 12) &
    (silver_df["kcal_flag"] == "normal_kcal") &
    (silver_df['age_group'] != "0-2")
].copy()
```

### Gold layer for exploartion data

```python
gold_df = quality_silver_data.copy()
```

```python
gold_df
```

The minimum recommended sodium intake for an average adult is 1500-2000 mg per day.
Rocmmendadtion for fiber intake for an average adult is 20-40g per day

```python
# Doing mask for high, low, normal sodium as 0 or 1
gold_df['is_low_sodium'] = ((gold_df["DR1TSODI"] < 1500)).astype(int)
gold_df['is_high_sodium'] = (gold_df["DR1TSODI"] > 2000).astype(int)

# Normal sodium is between 1500-2000
gold_df['is_normal_sodium'] = (((gold_df["DR1TSODI"] >= 1500) & (gold_df["DR1TSODI"] <= 2000))).astype(int)


# Doing mask for high, low, normal fiber as 0 or 1
gold_df['is_low_fiber'] = ((gold_df["DR1TFIBE"] < 20)).astype(int)
gold_df['is_high_fiber'] = ((gold_df["DR1TFIBE"] > 40)).astype(int)

# Normal fiber is between 20-40 
gold_df['is_normal_fiber'] = (((gold_df["DR1TFIBE"] >= 20) & (gold_df["DR1TFIBE"] <= 40))).astype(int)
```

### Saving gold_df to db

```python
engine = create_engine(
    "postgresql://postgres:postgres@localhost:5432/diploma_db"
)

gold_df.to_sql(
    "nhanes_gold",
    engine,
    index=False,
    if_exists="replace"
)
```

### Statistic calculation

```python
gold_cohort_stats_2 = pd.DataFrame([{
    "total_population": gold_df["SEQN"].count(),

    "median_sodium_mg": gold_df["DR1TSODI"].median(),
    "p90_sodium_mg": gold_df["DR1TSODI"].quantile(0.90),
    "pct_low_sodium": gold_df["is_low_sodium"].mean() * 100,
    "pct_normal_sodium": gold_df["is_normal_sodium"].mean() * 100,
    "pct_high_sodium": gold_df["is_high_sodium"].mean() * 100,

    "median_fiber_g": gold_df["DR1TFIBE"].median(),
    "p90_fiber_g": gold_df["DR1TFIBE"].quantile(0.90),
    "pct_low_fiber": gold_df["is_low_fiber"].mean() * 100,
    "pct_normal_fiber": gold_df["is_normal_fiber"].mean() * 100,
    "pct_high_fiber": gold_df["is_high_fiber"].mean() * 100
}])

gold_cohort_stats_2 = gold_cohort_stats_2.round(1)
gold_cohort_stats_2
```

```python
gold_cohort_stats_2.to_sql(
    "gold_stats",
    engine,
    index=False,
    if_exists="replace"
)
```
